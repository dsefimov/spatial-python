## Задача 2

**Описание**: Прочитайте файл heights.txt и напечатайте амплитуду значений в нем (разницу между самым большим и маленьким значением).

**Задание**: Самостоятельно напишите код, чтобы он соответствовал описанию

**Пример данных для проверки**:

```
Output: 9
```

````python
with open('./heights.txt') as f:
    values = tuple(map(int, f.read().split()))
    amplitude = max(values) - min(values)

print(amplitude)
````

````python
with open('./heights.txt') as f:

    max_value = 0
    min_value = float('inf')

    for value in f.read().split():
        value = int(value)

        max_value = max((max_value, value))
        min_value = min((min_value, value))

    amplitude = max_value - min_value

print(amplitude)
````

## Задача 3

**Описание**: Создайте новый файл depths.txt, содержащий значения полученные следующим образом: 10-x, x - соответсвующее значение из файла heights.txt

**Задание**: Самостоятельно напишите код, чтобы он соответствовал описанию

**Пример данных для проверки**:

```
Содержимое файла:
9 8 5 7 6
4 5 6 3 8
4 1 5 7 9
10 6 4 3 2
```

````python
with open('./heights.txt') as source_f, open('./depths.txt', 'w') as dest_f:
    for row in source_f:
        values = (str(10 - x) for x in map(int, row.split()))
        new_row = ' '.join(values)
        dest_f.write(new_row)
        dest_f.write('\n')
````

## Задача 4

**Описание**: Обновите сещуствующий файл depths.txt, заменив все значения меньше 5 на 0.

**Задание**: Самостоятельно напишите код, чтобы он соответствовал описанию

**Пример данных для проверки**:

```
Содержимое файла:
9 8 5 7 6
0 5 6 0 8
0 0 5 7 9
10 6 0 0 0
```

````python
path = r'./depths.txt'

# Открываем файл для чтения
with open(path, 'r') as file:
    # Считываем данные из файла
    lines = file.readlines()

# Преобразуем строки в числа и заменяем значения меньше 5 на 0
new_lines = []
for line in lines:
    values = [int(value) if int(value) >= 5 else 0 for value in line.split()]
    new_lines.append(' '.join(map(str, values)) + '\n')

# Открываем файл для записи и записываем изменённые данные
with open(path, 'w') as file:
    file.writelines(new_lines)
````

## Задача 5

**Описание**: Напишите функцию, которая принимает на вход путь к директории и расширение файла. Функция должна вернуть список путей к файлам с указанным расширением из директории

**Задание**: Самостоятельно напишите код, чтобы он соответствовал описанию

````python
import os

def get_files(folder, extension):
    return [os.path.join(folder, file) for file in os.listdir(folder) if file.endswith(extension)]


files = get_files('./', '.txt')
print(*files, sep='\n')
````

````python
from pathlib import Path

def get_files(folder, extension):
    folder = Path(folder)
    return [file.absolute() for file in folder.iterdir() if file.suffix == extension]


files = get_files('./', '.txt')
print(*files, sep='\n')
````

## Задача 5

**Описание**: Напишите функцию, которая принимает на вход путь к директории и расширение файла. Функция должна вернуть список путей к файлам с указанным расширением из самой дирекотрии и вложенных в неё директорий

````python
import os

def get_files(folder, extension):
    filtered_files = []
    for root, folders, files in os.walk(folder):
        filtered_files_part = [os.path.join(root, file) for file in files if file.endswith(extension)]
        filtered_files.extend(filtered_files_part)

    return filtered_files


files = get_files('./', '.ipynb')
print(*files, sep='\n')
````
