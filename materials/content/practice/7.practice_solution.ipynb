## Задача 2

**Описание**: Откройте файл countries.geojson и найдите страну с самым большим населением

**Задание**: Самостоятельно напишите код, чтобы он соответствовал описанию

**Пример данных для проверки**:

```
Output: China
```

````python
import json

with open('./countries.geojson') as f:
    data = json.load(f)

    max_population = 0
    country_with_max_population = None
    for feature in data["features"]:
        population = feature["properties"]["pop_est"]
        if population > max_population:
            max_population = population
            country_with_max_population = feature["properties"]["name"]

print(country_with_max_population)
````

## Задача 3

**Описание**: Откройте файл countries.geojson и посчитайте население европейских стран

**Задание**: Самостоятельно напишите код, чтобы он соответствовал описанию

**Пример данных для проверки**:

```
Output: 746398461
```

````python
import json

with open('./countries.geojson') as f:
    data = json.load(f)

    europe_population = 0
    for feature in data["features"]:
        if feature["properties"]["continent"] == "Europe":
            europe_population += feature["properties"]["pop_est"]

print(europe_population)
````

## Задача 4

**Описание**: Откройте файл countries.geojson и посчитайте суммарное население для каждого континента

**Задание**: Самостоятельно напишите код, чтобы он соответствовал описанию

**Пример данных для проверки**:

```
Output: {
    'Oceania': 36782844,
    'Africa': 1219176238,
    'North America': 573042112,
    'Asia': 4389144868,
    'South America': 418540749,
    'Europe': 746398461,
    'Seven seas (open ocean)': 140,
    'Antarctica': 4050
}
```

````python
import json

with open('./countries.geojson') as f:
    data = json.load(f)

    continents = {}

    for feature in data["features"]:
        population = feature["properties"]["pop_est"]
        continent = feature["properties"]["continent"]
        continents[continent] = continents.setdefault(continent, 0) + population

print(continents)
````

## Задача 5
**Описание**: Откройте файл countries.geojson и сохраните европейские страны из него, как новый geojson файл

**Задание**: Самостоятельно напишите код, чтобы он соответствовал описанию

````python
import json


with open('./countries.geojson') as f:
    data = json.load(f)
    data["features"] = [
        feature for feature in data["features"]
        if feature["properties"]["continent"] == 'Europe'
    ]

with open('./europe.geojson', 'w') as nf:
    json.dump(data, nf)
````
